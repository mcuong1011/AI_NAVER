Cau 2: Dien cac gia tri bi thieu

```python
df1 = data.copy()
display(df1.isna().sum())
display(df1)
```

* Phuong phap trung binh

```python
df1['sepal.length'].fillna(round(data['sepal.length'].mean()),inplace=True)
df1['sepal.width'].fillna(round(data['sepal.width'].mean()),inplace=True)
df1['petal.length'].fillna(round(data['petal.length'].mean()),inplace=True)
df1['petal.width'].fillna(round(data['petal.width'].mean()),inplace=True)
display(df1)
```

* Phuong phap tan so cao

```python
df1['variety'].fillna(df1['variety'].mode()[0], inplace=True)
display(df1)
```

```python
df1.isna().sum()
```

Cau 4: Chuan hoa du lieu min max

```python
df3 = df1.copy()

sepal_length_min = df3['sepal.length'].min()
sepal_length_max = df3['sepal.length'].max()

sepal_width_min = df3['sepal.width'].min()
sepal_width_max = df3['sepal.width'].max()

petal_length_min = df3['petal.length'].min()
petal_length_max = df3['petal.length'].max()

petal_width_min = df3['petal.width'].min()
petal_width_max = df3['petal.width'].max()
```

```python
df3['sepal.length'] = (df3['sepal.length'] - sepal_length_min) / (sepal_length_max - sepal_length_min)
df3['sepal.width'] = (df3['sepal.width'] - sepal_width_min) / (sepal_width_max - sepal_width_min)
df3['petal.length'] = (df3['petal.length'] - petal_length_min) / (petal_length_max - petal_length_min)
df3['petal.width'] = (df3['petal.width'] - petal_width_min) / (petal_width_max - petal_width_min)
display(df3)
```

Cau 5: Lien tuc hoa du lieu Variety

```python
df4 = df3.copy()
df4['variety'].unique()
```

```python
dummies = pd.get_dummies(df4['variety'])
display(dummies)
```

```python
df4 = pd.concat([df4,dummies], axis=1)
df4.drop('variety', axis=1, inplace=True)
display(df4)
```

Cau 6: Xay dung mo hinh hoi quy tuyen tinh 

```python
df5 = df1.copy()
```

```python
x = df5.drop('variety', axis = 1)
y = df5['variety']
tp = int(0.9 * len(x))
```

```python
x_train = x[:tp]
x_test = x[tp:]
y_train = y[:tp]
y_test = y[tp:]
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression().fit(x_train, y_train)
logreg.score(x_test, y_test)
```

```python
pred = logreg.predict(x_test)
print("Predicted label: {}".format(pred))
print("True label: {}".format(y_test.values))
```

Cau 9: Phan cum du lieu va loai bo ngoai lai

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = df4.copy()
x = x.values
X_petal = x[:,2:4]
kmeans = KMeans(n_clusters=3, random_state=0).fit(X_petal)
```

```python
labels = kmeans.labels_
cluster0 = X_petal[labels == 0]
cluster1 = X_petal[labels == 1]
cluster2 = X_petal[labels == 2]
```
